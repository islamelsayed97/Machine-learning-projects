### Model Building

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
def train(x,y):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
    clf = LogisticRegression()
    clf.fit(x_train,y_train)
    return round(clf.score(x_test,y_test)*100,2)
```

```python
train(x,y)
```

### StopWords Removal

```python
sw        = open("en_stopwords.txt").read().split('\n')
def remove_sw(text):
    words = text.lower().split()
    wanted_words=[]
    for word in words:
        if word not in sw:
            wanted_words.append(word)
            
    return ' '.join(wanted_words)
```

```python
remove_sw("I have a bad dog and another good dog")
```

```python
sw        = open("en_stopwords.txt").read().split('\n')
def remove_sw2(text):
    words = text.lower().split()
    return set(words).difference(set(sw))
```

```python
remove_sw2("I have a bad dog and another good dog")
```

```python
df['clean_text']=df.Text.apply(remove_sw)
df.head()
```

```python
x=cv.fit_transform(df.clean_text).toarray()
```

```python
train(x,y)
```

### Stemming

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
ps.stem("playing")
```

```python
ps.stem("bed")
```

```python
def stem(text):
    words = text.lower().split()
    stemmed_words = []
    for word in words:
        stemmed_words.append(ps.stem(word))
    return ' '.join(stemmed_words)
```

```python
stem("Hopefully will be able to get my iPhone 6s tod...")
```

# Arabic

```python
import codecs
from nltk.stem import ISRIStemmer
sw = codecs.open('stopWords.csv','r',encoding='utf-8').read().split('\n') ## For Arabic
isri = ISRIStemmer()                                                      ## For Arabic
def clean(text):
    text         = re.sub('[^-0-9]',' ',text)                        ## For Arabic
    words        = text.split()
    wanted_words =[]
    for word in words:
        if word not in sw:
            wanted_words.append(isri.stem(word))                          ## For Arabic
            
    return ' '.join(wanted_words)
```

```python
clean("      28 ")
```

```python
df = pd.read_csv('dataset.csv',names=['Text','label'])
df.head()
```

```python
df['clean_text'] = df.Text.apply(clean)
df.head()
```

```python
df.shape
```

```python
x = cv.fit_transform(df.clean_text).toarray()
y = df.label
```

```python
train(x,y)
```

```python
x = tfidf.fit_transform(df.clean_text)
x
```

```python
train(x,y)
```
